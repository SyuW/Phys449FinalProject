```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
def plot_result(N, folder_path):
    symm = ['True', 'False']
    color1 = ['r', 'b']
    plt.figure(figsize=(10,6))
    for i in range(len(symm)):
        data = np.load(folder_path+'/N='+str(N)+'/infidelity_N_'+str(N)+'_symm_'+symm[i]+'.npy')
        plt.plot(np.arange(2000)[::30], data[::30], marker = "v", markevery=1,color=color1[i])
    
    plt.legend(['U(1)-RNN', 'RNN'])
    plt.xlabel("Epochs")
    plt.ylabel("1-F")
    plt.title("N="+str(N))
    plt.yscale('log')
    fig = plt.show()
    return fig
```

```python
plot_result(4, 'final_results')
```

```python
plot_result(10, 'final_results')
```
