```python
import matplotlib.pyplot as plt
import numpy as np

import sys
import os

sys.path.append("C:\\Users\\Sam Yu\\Documents\\GitHub\\Phys449FinalProject")
```

```python
# create figure 4 from the paper
# energies and fidelities for N=4,10

Ns = [4, 10]
final_results_path = "C:\\Users\\Sam Yu\\Documents\\GitHub\\Phys449FinalProject\\final_results"
for _path_ in os.listdir(final_results_path):
    print(_path_)
```

```python
def create_energy_plot(num_spins, in_path, fig_no):
    
    data_path = os.path.join(in_path, f"N={num_spins}")
    
    # load the energy values from array
    energy_diff_symm_false = np.load(os.path.join(data_path, f"energy_diff_{num_spins}_symm_False.npy"))
    energy_diff_symm_true = np.load(os.path.join(data_path, f"energy_diff_{num_spins}_symm_True.npy"))
    epochs = np.arange(len(energy_diff_symm_false)) + 1
    
    with plt.ioff():
        fig, ax = plt.subplots()
    
    # plot everything
    nth = 19
    if fig_no == 4:
        ax.semilogy(epochs[::nth], energy_diff_symm_false[::nth], marker=">", label="RNN")
        ax.semilogy(epochs[::nth], energy_diff_symm_true[::nth], marker="<", label="U(1)-RNN")
        
    elif fig_no == 5:
        ax.plot(epochs[::nth], energy_diff_symm_false[::nth], marker=">", label="RNN")
        ax.plot(epochs[::nth], energy_diff_symm_true[::nth], marker="<", label="U(1)-RNN")
        ax.set_ylim([0.000, 0.06])
    else:
        pass
        
    ax.set_xlabel("Epochs")
    ax.set_ylabel(r"Energy difference, $\epsilon$")
    ax.set_title(f"Energy difference during training (N = {num_spins})")
    ax.legend()
    
    return fig
```

```python
efig_n_4 = create_energy_plot(num_spins=4, in_path=final_results_path, fig_no=4)
plt.show(efig_n_4)

plt.close()
```
